# Color Identification

```python
#FUNCTION THAT CONVERT HEX TO RGB
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))


#FUNCTION THAT HELP US GET AN IMAGE INTO PYTHON IN THE RGB SPACE
def get_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
def get_colors(image, number_of_colors, show_chart):
    # We resize the image for reduce time needed to extract the colors from the image
    modified_image = cv2.resize(image, (600, 400), interpolation = cv2.INTER_AREA)
    modified_image = modified_image.reshape(modified_image.shape[0]*modified_image.shape[1], 3)

    # kmeans create cluster of colors and these clusters will be our top color
    clf = KMeans(n_clusters = number_of_colors)
    # we extract the predicts about the same image
    labels = clf.fit_predict(modified_image)



    counts = Counter(labels)
    #find the colors clustered
    center_colors = clf.cluster_centers_
    # We get ordered colors by iterating through the keys
    ordered_colors = [center_colors[i]/255 for i in counts.keys()]
    # We multiply by 255 for get all colors again
    hex_colors = [RGB2HEX(ordered_colors[i]*255) for i in counts.keys()]
    rgb_colors = [ordered_colors[i]*255 for i in counts.keys()]

    if (show_chart):
        plt.figure(figsize = (8, 6))
        # put the colors (rgb) at the graph and the label in hex
        plt.pie(counts.values(), labels = hex_colors, colors = ordered_colors)

    return rgb_colors
```

# Serch images using color

```python
# Define some variables
IMAGE_DIRECTORY = 'images'
COLORS = {
    'GREEN': [0, 128, 0],
    'BLUE': [0, 0, 128],
    'YELLOW': [255, 255, 0]
}
images = []

for file in os.listdir(IMAGE_DIRECTORY):
    if not file.startswith('.'):
        images.append(get_image(os.path.join(IMAGE_DIRECTORY, file)))
```

##### Showing what images we have

```python
plt.figure(figsize=(20, 10))
for i in range(len(images)):
    plt.subplot(1, len(images), i+1)
    plt.imshow(images[i])
```

```python
# threshold define how different color of the image and selected color. 
# If threshold is too high we seeing blue images, if is too low the green maybe not match dark green or light green
def match_image_by_color(image, color, threshold = 60, number_of_colors = 10): 
    
    # first we extract primary colors 
    image_colors = get_colors(image, number_of_colors, False)
    # transform the color to a format we can compare
    selected_color = rgb2lab(np.uint8(np.asarray([[color]])))

    select_image = False
    for i in range(number_of_colors):
        curr_color = rgb2lab(np.uint8(np.asarray([[image_colors[i]]])))
        diff = deltaE_cie76(selected_color, curr_color)
        if (diff < threshold):
            select_image = True
    
    return select_image
```

```python
# Iterates all images and call our function for filter the image
def show_selected_images(images, color, threshold, colors_to_match):
    index = 1
    
    for i in range(len(images)):
        selected = match_image_by_color(images[i], color, threshold, colors_to_match)
        if (selected):
            plt.subplot(1, 5, index)
            plt.imshow(images[i])
            index += 1
```
